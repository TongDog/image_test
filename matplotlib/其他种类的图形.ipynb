## 1.散点图

```python
import matplotlib.pyplot as plt
import numpy as np

n = 1024
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)

T = np.arctan2(y,x)  # 颜色的值的确认
plt.scatter(y,x,s=75,c=T,alpha=0.5) # s为大小；c为颜色，会与每个点一一对应

plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))

plt.xticks(())# 隐藏索引ticks
plt.yticks(())

plt.show()
```

## 2.柱状图

```python
import matplotlib.pyplot as plt
import numpy as np

n = 12
x = np.arange(n)
y1 = (1-x/float(n)) * np.random.uniform(0.5,1.0,size=n)
y2 = (1-x/float(n)) * np.random.uniform(0.5,1.0,size=n)

plt.bar(x,+y1,facecolor='#66ccff',alpha=0.5,edgecolor='white')
plt.bar(x,-y2,facecolor='red',alpha=0.5,edgecolor='white')

for x_,y in zip(x,y1):
    plt.text(x_+0.04,y+0.05,'%.2f' % y,ha='center',va='bottom') # ha:横向对齐 # va ：纵向对齐

for x_,y in zip(x,y2):
    plt.text(x_+0.04,-y-0.05,'%.2f' % y,ha='center',va='top')
    
plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(np.arange(-1,1,0.2))
    
plt.show()
```

# 3.等高线图

meshgrid被用来创建一个二维网格

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    return (1-x/2 + x**5 + y**3 ) * np.exp(-x**2-y**2) # 计算高度的值，不用理解

n=256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)



X,Y = np.meshgrid(x,y)
# contour：等高线图。f:fill颜色 cmap：hot为暖色调，cool为冷色调
# 8：等高线的数目
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap=plt.cm.cool) 

C = plt.contour(X,Y,f(X,Y),8,colors='k',linewidth=0.5)
plt.clabel(C,inline=1,fontsize=10) # 绘制等高线的label 可以设置inline属性，让线穿过数字(True为不穿过)

plt.show()
# print(X,Y)
```

```python
# %matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D



fig = plt.figure()
ax = Axes3D(fig)

X  = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
# print(X,Y)
R = np.sqrt(X**2+Y**2)

Z = np.sin(R)

# rstride:横向分割颜色的跨度
# cstride:纵向
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

# 投影，zdir='z'指从z轴压下去，投影到xOy轴
# ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap=plt.get_cmap('rainbow'))
ax.set_zlim(-2,2)

plt.show()
# ax.imshow()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(figsize=(10,6))

# 创建一个3D子图
ax = fig.add_subplot(111,projection='3d')

u = np.linspace(0,2*np.pi,100) # u是0到2pi之间的100个点
v = np.linspace(0,np.pi,100)# v是0到pi之间的100个点
x = 10* np.outer(np.cos(u),np.sin(v))
y = 10* np.outer(np.sin(u),np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)),np.cos(v))

X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)

R = np.sqrt(X**2+Y**2)

Z = np.sin(R)


# 绘制3D曲面图
surf = ax.plot_surface(X,Y,Z,cmap='rainbow',edgecolor='none')

# 添加颜色条
fig.colorbar(surf,shrink=0.5,aspect=5)

ax.legend(['Surface'])
ax.view_init(elev=10,azim=125)

ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap=plt.get_cmap('rainbow'))
ax.set_zlim(-2,2)

plt.show()
```
