## figure的使用
可以声明多个图片

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x,y1)

plt.figure()
plt.plot(x,y2)

plt.show()
```

```python
# 或者可以给figure参数,下面为指定fig的编号和大小的例子：
plt.figure(num=114,figsize=(10,10)) # 编号会显示在窗口上，jupyter没窗口
plt.plot(x,y1,x,y2)
```

```python
# 可以调整颜色和其他参数
plt.figure(num=3,figsize=(5,5))
plt.plot(x,y1)
plt.plot(x,y2,color='red',linewidth=2,linestyle='--')
```

## 修改边框和坐标轴

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=3,figsize=(5,5))
plt.plot(x,y1)
plt.plot(x,y2,color='red',linewidth=2,linestyle='--')

# 设置取值范围
plt.xlim((-1,2))
plt.ylim((-2,3))

# 添加x轴与y轴的标记
plt.xlabel("x label")
plt.ylabel("y label") 

# 修改某个轴的显示精度
new_ticks = np.linspace(-1,2,5) # -1到2之间5个点
plt.xticks(new_ticks)
# 也可以修改特定数值的特定标签
# 记得数量要对应好，不然会失效
# 也开修改字体，使用$really\ good$，空格要加转字符"\ " $$之间表示为正则表达式(Latex)
plt.yticks(
    [-2,-1.8,-1,1.22,3],
    [r'$really\ bad$',r'$bad\  \alpha$','normal','good','really good']
)

# 修改ax,中文为”脊梁“，指上下左右四个边框
ax = plt.gca()
ax.spines['right'].set_color('none') # 消失
ax.spines['top'].set_color('none')


# 用坐标轴当作脊梁
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', -1))
ax.spines['left'].set_position(('data', 0))
plt.show()
```

## legend图例

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=3,figsize=(5,5))
plt.plot(x,y1)
plt.plot(x,y2,color='red',linewidth=2,linestyle='--')

# 设置取值范围
plt.xlim((-1,2))
plt.ylim((-2,3))

# 添加x轴与y轴的标记
plt.xlabel("x label")
plt.ylabel("y label") 

# 修改某个轴的显示精度
new_ticks = np.linspace(-1,2,5) # -1到2之间5个点
plt.xticks(new_ticks)
# 也可以修改特定数值的特定标签
# 记得数量要对应好，不然会失效
# 也开修改字体，使用$really\ good$，空格要加转字符"\ " $$之间表示为正则表达式(Latex)
plt.yticks(
    [-2,-1.8,-1,1.22,3],
    [r'$really\ bad$',r'$bad\  \alpha$','normal','good','really good']
)

plt.plot(x,y2,label='up')
plt.plot(x,y1,label='down',color='red',linewidth=2,linestyle='--')

# 设置图例
plt.legend() # 不加参数的话为默认图例

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=3,figsize=(5,5))
plt.plot(x,y1)
plt.plot(x,y2,color='red',linewidth=2,linestyle='--')

# 设置取值范围
plt.xlim((-1,2))
plt.ylim((-2,3))

# 添加x轴与y轴的标记
plt.xlabel("x label")
plt.ylabel("y label") 

# 修改某个轴的显示精度
new_ticks = np.linspace(-1,2,5) # -1到2之间5个点
plt.xticks(new_ticks)
# 也可以修改特定数值的特定标签
# 记得数量要对应好，不然会失效
# 也开修改字体，使用$really\ good$，空格要加转字符"\ " $$之间表示为正则表达式(Latex)
plt.yticks(
    [-2,-1.8,-1,1.22,3],
    [r'$really\ bad$',r'$bad\  \alpha$','normal','good','really good']
)

l1, = plt.plot(x,y2,label='up')
l2, = plt.plot(x,y1,label='down',color='red',linewidth=2,linestyle='--')

# 设置图例
# handle:选择放入legend的点。注意传入lengend必须为一个元组，在l1后加一个“，”
# label：可以指定不同的名字
# loc为best会自动挑选一个好的位置来放图例,调整窗口大小时也会生效。还有其他参数，可以把图例放在其他位置
plt.legend(handles=[l1,l2],labels=['aa','asdadsa'],loc='best')  

plt.show()
```

```python
# 可以只打印出一条线，把handles中的l2去掉就行
plt.legend(handles=[l1,],labels=['aa'],loc='best')  
```

## 标注

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = x + 1
y2 = x**2

plt.figure(num=3,figsize=(5,5))
plt.plot(x,y1)
# plt.plot(x,y2,color='red',linewidth=2,linestyle='--')

# 设置取值范围
plt.xlim((-1,2))
plt.ylim((-2,3))

# 修改ax,中文为”脊梁“，指上下左右四个边框
ax = plt.gca()
ax.spines['right'].set_color('none') # 消失
ax.spines['top'].set_color('none')
# 用坐标轴当作脊梁
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

x0 = 1
y0 = x0+1

# scatter放置一个点
plt.scatter(x0,y0,color='red',s=50)

# [x0,x0],[y0,0]表示画x轴x0画到x0，y轴从0画到y0的一条直线
plt.plot([x0,x0],[y0,0],'k--',lw=2.5) # k--代表使用黑色的虚线 lw:line weights

# 加标注
# xy参数指坐标从哪开始
# xytext：文字描述,参数指横坐标加30，纵坐标减30打印点
plt.annotate(r'$2x+1 = %s$' % y0, xy=(x0,y0),xycoords='data',xytext=(+30,-30),
             textcoords='offset points',color='red',# offect point:基于xy的点进行xtext的内容
             arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2',color='red',)# 线的性质
             # connectionstyle指弧度，arrowstyle：箭头样式
             )

# 也可以加文字
plt.text(0,1,r'$2x+1 = %s$' % y0,fontdict={'color':'blue','fontsize':10},)

plt.show()
```

## 可见度调整

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = x + 1
y2 = x**2

plt.figure(num=3,figsize=(5,5))
plt.plot(x,y1)
# plt.plot(x,y2,color='red',linewidth=2,linestyle='--')

# 设置取值范围
plt.xlim((-1,2))
plt.ylim((-2,3))

# 修改ax,中文为”脊梁“，指上下左右四个边框
ax = plt.gca()
ax.spines['right'].set_color('none') # 消失
ax.spines['top'].set_color('none')
# 用坐标轴当作脊梁
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    # set_bbox:设置坐标轴label的背景颜色
    label.set_bbox(dict(facecolor='blue', edgecolor='white', alpha=0.7)) # alpha：透明度
    
plt.show()
```
