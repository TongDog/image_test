# 多合一显示
plt.subplot 和 plt.subplots 是两个用于创建和操纵 Matplotlib 子图的不同函数。

plt.subplot:
* 用于创建单个子图。
* 接受三个参数：行数 (nrows)、列数 (ncols) 和子图的编号 (num)。
* 返回一个 Axes 对象，代表创建的子图。<br/>

plt.subplots:
* 用于创建一个包含多个子图的 Figure 对象。
* 接受多个参数，包括 nrows (行数)、ncols (列数) 和 figsize (图的大小)。
* 返回两个对象：一个 Figure 对象和一组 Axes 对象数组，代表创建的子图。
简单来说，plt.subplot 用于创建单个子图，而 plt.subplots 用于创建一个包含多个子图的 Figure 对象。在处理多个子图时，plt.subplots 通常更为方便。

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

x = np.arange(0, 10, 0.1)
y = x**2

plt.subplot(2,2,1) # 在第一个位置2行2列
plt.xlim([0,10])
plt.plot(x,y) # x轴和y轴

plt.subplot(2,2,2)
plt.xlim([0,4*np.pi])
plt.plot(x,np.sin(x))

plt.subplot(2,2,3) 
plt.plot([0,1],[0,3])

plt.subplot(2,2,4) 
plt.plot([0,1],[0,4])

plt.show()
```

## 换一个分的方式


```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

x = np.arange(0, 10, 0.1)
y = x**2

plt.subplot(2,1,1) # 修改这里即可
plt.xlim([0,10])
plt.plot(x,y) # x轴和y轴

plt.subplot(2,3,4) # 这里没懂怎么分
plt.xlim([0,4*np.pi])
plt.plot(x,np.sin(x))

plt.subplot(2,3,5) 
plt.plot([0,1],[0,3])

plt.subplot(2,3,6) 
plt.plot([0,1],[0,4])

plt.show()
```

## 另一种分法

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# 1.
plt.figure()
 # (3,3)创建一个3*3的空间
 # (0,0)该张图像从0,0开始排布
 # colspan、rowspan:该行列的跨度，3，1指3列1行
ax1 = plt.subplot2grid((3,3),(0,0),colspan=3,rowspan=1)
ax1.plot([1,2],[1,2])
ax1.set_title('ax1')

ax2 = plt.subplot2grid((3,3),(1,0),colspan=2)
ax2.plot([1,2],[1,2])

ax3 = plt.subplot2grid((3,3),(2,0),rowspan=1)
ax4 = plt.subplot2grid((3,3),(2,1),rowspan=1)

ax5 = plt.subplot2grid((3,3),(1,2),rowspan=2) 
```

```python
# 2
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.figure()
gs = gridspec.GridSpec(3,3)
ax1 = plt.subplot(gs[0,:])
ax2 = plt.subplot(gs[1,:2])
ax3 = plt.subplot(gs[1:,2])
ax4 = plt.subplot(gs[-1,0])
ax5 = plt.subplot(gs[-1,-2])
```
