```python
from matplotlib import pyplot as plt
import numpy as np

# 定义正弦波函数的参数
A = 100  # 振幅
B = 2 * np.pi / 1000  # 频率，使得1公里内有一个完整的周期
C = 0  # 相位移动
D = 100  # 垂直位移，确保y轴的值在0到200之间

x = np.linspace(0, 1000, num=1000*4, endpoint=True)
# print(x)
# 应用正弦波函数
y = A * np.sin(B * x + C) + D

# 绘制散点图
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=1)
plt.title("正弦波散点图")
plt.xlabel("里程数 (米)")
plt.ylabel("参数值")
plt.grid(True)
plt.show()

# 返回数据点
x, y
```
